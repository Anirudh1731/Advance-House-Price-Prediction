```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/train.csv')
```

# **Dealing Categorical NAN values**

```python
categorical_nan=[feature for feature in data.columns if data[feature].dtypes=='O' and data[feature].isnull().sum()>1]
```

```python
categorical_nan
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

data=replace_cat_feature(data,categorical_nan)

data[categorical_nan].isnull().sum()
```

```python
data.head()
```

```python
data[categorical_nan].isnull().sum()
```

# **Now to handle numerical values**

```python
numerical_values=[feature for feature in data.columns if data[feature].dtypes!='O' and data[feature].isnull().sum()>1]
```

```python
numerical_values
```

```python
for feature in numerical_values:
  median_val=data[feature].median()

  data[feature+'nan']=np.where(data[feature].isnull(),1,0)
  data[feature].fillna(median_val,inplace=True)



```

```python
data[numerical_values].isnull().sum()
```

# **Handling datetime**

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
  data[feature]=data['YrSold']-data[feature]
```

```python
data[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
data.head()
```

# **Now making the features normally distributed after analyzing it during data analysis**

```python
for feature in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:

  data[feature]=np.log(data[feature])
```

```python
data.head()
```

# **Handling the rare catrgorical features **

```python
categorical_features=[feature for feature in data.columns if data[feature].dtypes=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
  '''here if it is rare and if it is only 1% of entire data then we are giving it a new name call rare'''
  temp=data.groupby(feature)['SalePrice'].count()/len(data)

  temp_id=temp[temp>0.01].index
  data[feature]=np.where(data[feature].isin(temp_id),data[feature],'Rare_var')
```

```python
temp
```

```python
data.head(50)
```

```python
for feature in categorical_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```
